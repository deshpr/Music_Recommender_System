```python
import pandas as pd
import numpy as np

# Read album information
df_album = pd.read_csv('.\\kaggle_songs.csv')
df_album.head()
df_album.drop(df_album.columns[[0, 1]], axis=1, inplace=True)
df_album.head()
```

```python
# Read the track information
df_tracks = pd.read_csv('.\\unique_tracks.csv',encoding = "ISO-8859-1")
df_tracks.head()
df_tracks.drop(df_tracks.columns[[0]], axis=1, inplace=True)
df_tracks.head()
```

```python
# Read the artist information
df_artist = pd.read_csv('.\\unique_artists.csv',encoding = "ISO-8859-1")
df_artist.drop(df_artist.columns[[0]], axis=1, inplace=True)
df_artist.head()
```

```python
# Read the evaluation file
df_playcnt = pd.read_csv('.\\kaggle_visible_evaluation_triplets.txt', sep="\t", header=None)
df_playcnt.columns = ["userid", "songid", "playcount"]

# Remove songs with play count greater than 3 (treat them as noise, since they are not very common)
df_playcnt = df_playcnt.loc[df_playcnt['playcount'] <= 3]
df_playcnt.head()
df_playcnt.shape
unique_users = df_playcnt['userid'].unique()
print("User count after reducing play count = {}".format(len(unique_users)))
#df_playcnt['playcount'].describe()
```

```python
catalog_prep=pd.merge(pd.merge(df_album, df_tracks, on='songid'),df_artist, on='artistname')
catalog_set= pd.merge(catalog_prep,df_playcnt, on='songid')
catalog_set.head()
```

```python
# Get filtered American artists
df_artist_filtered = pd.read_csv('.\\Filtered_Artists.csv',encoding = "ISO-8859-1", header=None)
df_artist_filtered.columns = ["artistname"]
df_artist_filtered.head()
```

```python
# Final Data Frame to work from (containing song, artist and the play count)
final_df= pd.merge(catalog_set,df_artist_filtered, on ='artistname')
print(final_df.shape)
final_df.head()
#temp = final_df_[final_df['songid'] == None]
```

```python
catalog_train = final_df.copy(deep = True)
catalog_train.to_csv("rahul_train", encoding='utf-8', index=False)
print("written_to_rahul_train")
```

```python
# Prepare the Item Matrix
item_matrix_prep=catalog_train.groupby(['songid','artistname'],as_index=False)
item_matrix_prep=item_matrix_prep.aggregate(np.sum)
item_matrix = pd.get_dummies(item_matrix_prep[['songid','artistname']], columns=['artistname'],prefix=[None])
item_matrix.shape
print("song matrix = {}".format(item_matrix.shape))
item_matrix.head()
#f = open('item_artist.txt','w')
#f.write(str(item_matrix.columns.values))
#f.close()
```

```python
# Prepare the User Matrix
user_matrix_prep=catalog_train.groupby(['userid','artistname'],as_index=False)
user_matrix_prep=user_matrix_prep.aggregate(np.sum)
user_matrix = pd.pivot_table(user_matrix_prep, values='playcount', index='userid', columns=['artistname'], aggfunc=np.sum,fill_value=0)
user_matrix.shape
print("User matrix = {}".format(user_matrix.shape))
user_matrix.head()

#f = open('user_artist.txt','w')
#f.write(str(user_matrix.columns.values))
#f.close()
```

```python
# Normalize User Play Count information
user_normalized= (user_matrix - user_matrix.values.min()) / (user_matrix.values.max() - user_matrix.values.min())
user_normalized.head()
user_normalized.columns
print("User matrix = {}".format(user_normalized.shape))
user_normalized.head()
user_normalized.to_csv("rahul_user_normalized.csv", encoding='utf-8', index=False)
```

```python
# This cell loads the cosine matrix from the file rahul_recommendation_final_2_top_all.csv. This cosine matrix was prepared
# as an offline process and written to this file since preparing the cosine matrix is a 4-5 hour process.

#df_output.head()
print("start")
recommendation = pd.DataFrame.from_csv("rahul_recommendation_final_2_top_all.csv", encoding='utf-8', index_col=None)
recommendation.head()
#print(recommendation.index)
#print(recommendation.columns.tolist())
#print(type(recommendation))
print("done")
```

```python
# This cell calculates the information needed to calculate the average recall. 
# To get the recall we need to know the entire list of songs that the user listened (the entire truth) and the list of songs we
# predict the user would listen (so the count of the values greater than 0.5)

user_recall = {}
print(recommendation.columns.tolist())
print(type(recommendation))

for index, entry in recommendation.iterrows():
    user_id = entry["userid"]
    song_id = entry["songid"]
#    print("at index = {}".format(index))
    cosine_value = entry["cosine"]
    if not user_id in user_recall:
        user_recall[user_id] = {'truth':0, 'prediction':0}
    if cosine_value != 0:
        user_recall[user_id]['truth'] += 1
        if cosine_value > 0.5:
            user_recall[user_id]['prediction'] += 1
#        else:
#            print("found something!: {}".format(user_id))
print("Calculated information for recall")
print(user_recall)
```

```python
# This cell also writes the information needed to calculate average user recall to a file since the process takes a lot of time
# because the user song combination is so large.
import json
with open('user_recall.txt', 'w') as file:
     file.write(json.dumps(user_recall)) # use `json.loads` to do the reverse
print("done")
```

```python
# This cell calculates the recall per user, and then uses that to calculate the average recall for all users.

print("Recall per user:")
total_recall = 0
for user_id, value in user_recall.items():
    if user_recall[user_id]['truth'] == 0:
        user_recall[user_id]['truth'] = 1
    recall = user_recall[user_id]['prediction']/user_recall[user_id]['truth']
#    print("recall = {}".format(recall * 100))
    total_recall += recall
print("Average User Recall = {} %".format(total_recall/len(user_recall) * 100))
```

```python
# To calculate the precision we would like to develop a dictionary where the key is the user id, and the value is another dictionary, 
# where the key is the song id, and the value is the cosine value between the user and that song.
# This makes precision calculations easier.

user_recommendation = {}
#print("prepare dictionary")
count_position = 0
for index, row in recommendation.iterrows():
    user_id = row["userid"]
    song_id = row["songid"]
    cosine = row["cosine"]
#    if index == 50000:
#        break
#    print("index = {}, user_id = {}, song id = {}".format(index, user_id, song_id))
    if not user_id in user_recommendation:
        user_recommendation[user_id] = {}
    # There is only one song per user record (we don't recommend same song twice)
    user_recommendation[user_id][song_id] = cosine

print("Prepared the dictionary")
print(user_recommendation)
```
